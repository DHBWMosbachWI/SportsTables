# Teaser Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

fontsize = 17
width = 0.45
data = {
    "VIZNET": {
        "textual_cols_avg": 1.34,
        "numerical_cols_avg": 0.19
    },
    "TURL": {
        "textual_cols_avg": 1.61,
        "numerical_cols_avg": 0,
    },
    "SemTab2019": {
        "textual_cols_avg": 1.57,
        "numerical_cols_avg": 0.01,
    },
    "SemTab2020": {
        "textual_cols_avg": 1.45,
        "numerical_cols_avg": 0.001,
    },
    "SemTab2021": {
        "textual_cols_avg": 2.77,
        "numerical_cols_avg": 1.09,
    },
    "GitTables": {
        "textual_cols_avg": 3.62,
        "numerical_cols_avg": 3.2,
    },
    "Enterprise Data Lakes": {
        "textual_cols_avg": 2.02,
        "numerical_cols_avg": 8.1
    }
}


fig, ax = plt.subplots(figsize=(20,6))

rects = []

for idx_corpora, corpora in enumerate(data.keys()):
    if corpora == "Enterprise Data Lakes":
        total_cols = data[corpora]["textual_cols_avg"]+data[corpora]["numerical_cols_avg"]
        rects.append(ax.bar([(idx_corpora+0.5)-width/2], [round(data[corpora]["textual_cols_avg"]/total_cols*100,2)], width=width, label=f"Textual Cols. per Table", color="C0", edgecolor="black"))
        rects.append(ax.bar([(idx_corpora+0.5)+width/2], [round(data[corpora]["numerical_cols_avg"]/total_cols*100,2)], width=width, label=f"Numerical Cols. per Table", color="C1", edgecolor="black"))
    else:
        total_cols = data[corpora]["textual_cols_avg"]+data[corpora]["numerical_cols_avg"]
        rects.append(ax.bar([idx_corpora-width/2], [round(data[corpora]["textual_cols_avg"]/total_cols*100,2)], width=width, label=f"Textual Cols. per Table", color="C0"))
        rects.append(ax.bar([idx_corpora+width/2], [round(data[corpora]["numerical_cols_avg"]/total_cols*100,2)], width=width, label=f"Numerical Cols. per Table", color="C1"))
    
    
for rect in rects:
    if rect == rects[len(rects)-1] or rect == rects[len(rects)-2]:
        ax.bar_label(rect, fmt="%.2f%%", fontsize=14, fontweight="bold")
    else:
        ax.bar_label(rect,fmt="%.2f%%", fontsize=14)
        
## annotations
ax.annotate("", xy=(-0.5,-11), xytext=(5.5,-11), arrowprops=dict(arrowstyle="|-|", facecolor="black", lw=1.5), annotation_clip=False)
ax.annotate('Existing semantically annotated Corporas',xy=(0.5,-15),xytext=(1,-17), annotation_clip=False, fontsize=17, weight="bold")#
ax.annotate("", xy=(6,-11), xytext=(7,-11), arrowprops=dict(arrowstyle="|-|", facecolor="black", lw=1.5), annotation_clip=False)
ax.annotate('Real-World',xy=(0.5,-0.275),xytext=(6.125,-17), annotation_clip=False, fontsize=17, weight="bold")

ax.set_xticks(list(range(0, len(data.keys())-1))+[6.5], list(data.keys()), fontsize=15)
sportsDB_xtick_label = ax.get_xticklabels()[-1]
sportsDB_xtick_label.set_fontweight("bold")

ax.set_ylabel("Avg. Perc. of Columns per Table", fontsize=17)

# Hide the right and top spines
ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)

ax.legend(handles=rects[:2], loc="upper center", bbox_to_anchor=(0.5,1.15), ncol=2, fancybox=True, shadow=True, fontsize=fontsize)
plt.savefig(f"./plots/Teaser.png", bbox_inches="tight", dpi=100, transparent=True)
plt.show()
```
